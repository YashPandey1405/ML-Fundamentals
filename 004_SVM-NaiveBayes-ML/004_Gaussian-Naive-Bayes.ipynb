# Variants of Naive Bayes Classifier

There are three main variants of the Naive Bayes classifier:

- **Bernoulli Naive Bayes**: Suitable for binary/Boolean features, often used in `NLP tasks like text classification`.
- **Multinomial Naive Bayes**: Ideal for count-based features (e.g., word frequencies), commonly used in `NLP for document classification`.
- **Gaussian Naive Bayes**: Applied when features follow a normal distribution. `This variant is demonstrated in this notebook`.

Each variant is chosen based on the nature of the data and problem.

```python
# Gaussian Naive Bayes Classifier.....

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

### Iris Dataset

The **Iris dataset** is a well-known dataset in machine learning and statistics, often used for classification tasks. It contains information about three different species of Iris flowers: **Setosa**, **Versicolor**, and **Virginica**.

### Dataset Details:
- **Features**: 4 attributes (or features) per flower:
  1. **Sepal Length** (cm)
  2. **Sepal Width** (cm)
  3. **Petal Length** (cm)
  4. **Petal Width** (cm)

- **Target Variable**: The species of the flower, which can be one of the following:
  - **Setosa** , **Versicolor** & **Virginica**

### Key Points:
- The dataset contains **150 samples**, with 50 samples of each species.
- The features are numeric, and the task is to **classify** the flowers based on these attributes.

```python
X, y = load_iris(return_X_y = True)
```

```python
#  X Has 4 features , That's Why 4 Values in One Row....
X
```

```python
# Target Variable y Has 3 Values (0,1,2)....
# 0 = Setosa, 1 = Versicolor, 2 = Virginica....
y
```

```python
# Splitting The Data Into Train And Test Set.....
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
X_train
```

```python
X_test
```

# Gaussian Naive Bayes Implementation......

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(X_train, y_train)
```

```python
# Predicting The Test Set....
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

# Accuracy Of The Model....
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")

# Accuracy: 0.64 i.e 64%....
```
